```python
import tkinter as tk
import numpy as np
import math as m
root = tk.Tk()
root.title("calculator")
root.geometry("300x400")
root.resizable(False , False)
entry = tk.Entry(root ,width = '20',font = ("Arial",15) , justify = "right")
entry.grid(row = 0 ,column = 0 ,sticky = 'w')

label = tk.Label(root , text = '',font = ("Arial",10) ) 
label.grid()
#function
def num():
    a = np.sqrt(float(entry.get()))
    entry.delete(0 ,tk.END)
    entry.insert(0,a)
def num1():
    b = np.square(float(entry.get()))
    entry.delete(0 ,tk.END)
    entry.insert(0,b)
def num2():
    c = np.sin(float(entry.get()))
    entry.delete(0,tk.END)
    entry.insert(0 , c)
def num3():
    d = np.cos(float(entry.get()))
    entry.delete(0,tk.END)
    entry.insert(0 , d)
def num4():
    e = np.tan(float(entry.get()))
    entry.delete(0,tk.END)
    entry.insert(0 , e)
def num5():
    f = np.log(float(entry.get()))
    entry.delete(0,tk.END)
    entry.insert(0 , f)
def num6():
    g = m.factorial(int(entry.get()))
    entry.delete(0,tk.END)
    entry.insert(0 , g)

def num7():
    entry.delete(0,tk.END)

def num8():
    entry.insert(tk.END ,"1")

def num9():
    entry.insert(tk.END ,"2")

def num10():
    entry.insert(tk.END ,"3")

def num11():
    entry.insert(tk.END ,"+")

def num12():
    entry.insert(tk.END ,"4")

def num13():
    entry.insert(tk.END ,"5")

def num14():
    entry.insert(tk.END ,"6")

def num15():
    entry.insert(tk.END ,"-")

def num16():
    entry.insert(tk.END ,"7")

def num17():
    entry.insert(tk.END ,"8")

def num18():
    entry.insert(tk.END ,"9")

def num19():
    entry.insert(tk.END ,"*")

def num20():
    entry.insert(tk.END ,"0")

def num21():
    entry.insert(tk.END ,".")

def num22():
    h = eval(entry.get())
    entry.delete(0,tk.END)
    entry.insert(0 , h)

def num23():
    entry.insert(tk.END ,"/")
    
def num24():
    i= np.array(eval(entry.get()))
    j = np.std(i)
    entry.delete(0, tk.END)
    entry.insert(0,j)

def num25():
    k = np.array(eval(entry.get()))
    l = np.mean(k)
    entry.delete(0, tk.END)
    entry.insert(0,l)

def num26():
    m = np.array(eval(entry.get()))
    n = np.var(m)
    entry.delete(0, tk.END)
    entry.insert(0,n)

def num27():
    o = np.array(eval(entry.get()))
    p = np.median(o)
    entry.delete(0, tk.END)
    entry.insert(0,p)

def num28():
    entry.insert(tk.END ,",")

button = tk.Button(root ,text = "√x" , command = num ,width=3 ,height =2 ,bg = "black" ,fg = "white" )
button.grid(row =2 , column = 0,pady = 5,sticky = 'w')

button = tk.Button(root ,text = "x²" , command = num1 ,width=3 ,height =2 ,bg = "black" ,fg = "white" )
button.grid(row =3 , column = 0,pady = 5,sticky = 'w')

button = tk.Button(root ,text = "sin" , command = num2 ,width=3 ,height =2 ,bg = "black" ,fg = "white")
button.grid(row =4 , column = 0,pady = 5,sticky = 'w')

button = tk.Button(root ,text = "cos" , command = num3 ,width=3 ,height =2,bg = "black" ,fg = "white")
button.grid(row =5 , column = 0,pady = 5,sticky = 'w')

button = tk.Button(root ,text = "tan" , command = num4 ,width=3 ,height =2 ,bg = "black" ,fg = "white")
button.grid(row =6 , column = 0,pady = 5,sticky = 'w')

button = tk.Button(root ,text = "log" , command = num5 ,width=3 ,height =2,bg = "black" ,fg = "white"  )
button.grid(row =1 , column = 0,pady = 0, padx = 50 ,sticky = 'w')

button = tk.Button(root ,text = "x!" , command = num6 ,width=3 ,height =2 ,fg = "black" ,bg = "white" )
button.grid(row =7 , column = 0,pady = 5,sticky = 'w')

button = tk.Button(root ,text = "AC" , command = num7 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =1 , column = 0,pady = 5,sticky = 'w')

button = tk.Button(root ,text = "1" , command = num8 ,width=3 ,height =2 ,bg = "black" ,fg = "white")
button.grid(row =2 , column = 0,pady = 0, padx = 50 ,sticky = 'w')

button = tk.Button(root ,text = "2" , command = num9 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =2 , column = 0,pady = 0, padx = 100 ,sticky = 'w')

button = tk.Button(root ,text = "3" , command = num10 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =2 , column = 0,pady = 0, padx = 150 ,sticky = 'w')

button = tk.Button(root ,text = "+" , command = num11 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =2 , column = 0,pady = 0, padx = 200 ,sticky = 'w')

button = tk.Button(root ,text = "4" , command = num12 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =3 , column = 0,pady = 0, padx = 50 ,sticky = 'w')

button = tk.Button(root ,text = "5" , command = num13 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =3 , column = 0,pady = 0, padx = 100 ,sticky = 'w')

button = tk.Button(root ,text = "6" , command = num14 ,width=3 ,height =2 ,bg = "black" ,fg = "white" )
button.grid(row =3 , column = 0,pady = 0, padx = 150 ,sticky = 'w')

button = tk.Button(root ,text = "-" , command = num15 ,width=3 ,height =2 ,fg = "black" ,bg = "white"   )
button.grid(row =3 , column = 0,pady = 0, padx = 200 ,sticky = 'w')

button = tk.Button(root ,text = "7" , command = num16 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =4 , column = 0,pady = 0, padx = 50 ,sticky = 'w')

button = tk.Button(root ,text = "8" , command = num17 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =4 , column = 0,pady = 0, padx = 100 ,sticky = 'w')

button = tk.Button(root ,text = "9" , command = num18 ,width=3 ,height =2 ,fg = "black" ,bg = "white"   )
button.grid(row =4 , column = 0,pady = 0, padx = 150 ,sticky = 'w')

button = tk.Button(root ,text = "X" , command = num19 ,width=3 ,height =2 ,fg = "black" ,bg = "white"   )
button.grid(row =4 , column = 0,pady = 0, padx = 200 ,sticky = 'w')

button = tk.Button(root ,text = "0" , command = num20 ,width=3 ,height =2 ,bg = "black" ,fg = "white"  )
button.grid(row =5 , column = 0,pady = 0, padx = 50 ,sticky = 'w')

button = tk.Button(root ,text = "." , command = num21 ,width=3 ,height =2 ,fg = "black" ,bg = "white"   )
button.grid(row =5 , column = 0,pady = 0, padx = 100 ,sticky = 'w')

button = tk.Button(root ,text = "=" , command = num22 ,width=3 ,height =2 ,fg = "black" ,bg = "white"   )
button.grid(row =5 , column = 0,pady = 0, padx = 150 ,sticky = 'w')

button = tk.Button(root ,text = "/" , command = num23 ,width=3 ,height =2 ,fg = "black" ,bg = "white"  )
button.grid(row =5 , column = 0,pady = 0, padx = 200 ,sticky = 'w')

button = tk.Button(root ,text = "std" , command = num24 ,width=3 ,height =2  ,fg = "black" ,bg = "white"  )
button.grid(row =6 , column = 0,pady = 0, padx = 50 ,sticky = 'w')

button = tk.Button(root ,text = "mean" , command = num25 ,width=4 ,height =2  ,fg = "black" ,bg = "white"  )
button.grid(row =6 , column = 0,pady = 0, padx = 100 ,sticky = 'w')

button = tk.Button(root ,text = "variance" , command = num26 ,width=6 ,height =2 ,fg = "black" ,bg = "white"   )
button.grid(row =6 , column = 0,pady = 0, padx = 150 ,sticky = 'w')

button = tk.Button(root ,text = "median" , command = num27 ,width=6 ,height =2 ,fg = "black" ,bg = "white"   )
button.grid(row =7 , column = 0,pady = 0, padx = 50 ,sticky = 'w')

button = tk.Button(root ,text = "," , command = num28 ,width=3 ,height =2 ,bg = "black" ,fg = "white"   )
button.grid(row =1 , column = 0,pady = 0, padx = 100 ,sticky = 'w')

root.mainloop()
```
